### PHASE 1 — Business Objective (Write in Project)

Analyze sales pipeline data to identify revenue drivers, sales performance patterns, sector contribution, conversion efficiency, and regional trends to improve business growth strategy.

### PHASE 2 — Feature Engineering

#### Convert Dates
#converting the column from string (text) to datetime format

```python
merged_df['engage_date'] = pd.to_datetime(merged_df['engage_date'])
merged_df['close_date'] = pd.to_datetime(merged_df['close_date'])
merged_df
```

#### total revenue

```python
merged_df['revenue'] = merged_df['close_value'].sum()
```

```python
merged_df.head()
```

#### We get a numeric column to calculate conversion rate easily

```python
import numpy as np

merged_df['deal_won_flag'] = np.where(merged_df['deal_stage'] == 'Won', 1, 0)
merged_df
```

#### Create Closing Days
To measure how long each deal takes to close.

```python
merged_df['closing_days'] = (merged_df['close_date'] - merged_df['engage_date']).dt.days
merged_df
```

#### Extract Year & Month
checking sales trends over time.

```python
merged_df['close_year'] = merged_df['close_date'].dt.year
merged_df['close_month'] = merged_df['close_date'].dt.month
merged_df
```

#### Revenue Per Employee
which employee genrated more revenue

```python
merged_df['revenue_per_employee'] = merged_df['revenue'] / merged_df['employees']
merged_df
```

## PHASE 3 —  Revenue Analysis

##### Total Revenue


```python
merged_df['close_value'] = pd.to_numeric(merged_df['close_value'], errors='coerce')
```

```python
#its treating the close value as the string so we converting it into the float value
```

```python
merged_df['close_value'].fillna(0, inplace=True)
```

```python
revenue = merged_df['close_value'].sum()
print(revenue)
```

```python
merged_df['close_value'].dtype
```

#### product vise revenue

```python
merged_df.groupby("product")["close_value"].sum().reset_index().sort_values(by="close_value",ascending = False)
```

#### Revenue by Sector

```python
merged_df.groupby('sector')['close_value'].sum().reset_index().sort_values(by ='close_value', ascending=False)
```

#### Revenue by Sales Agent

```python
merged_df.groupby("sales_agent")["close_value"].sum().reset_index().sort_values(by ="close_value",ascending=False).head(10)
```

#### Revenue by Region

```python
merged_df.groupby('regional_office')['close_value'].sum().reset_index().sort_values(by = "close_value",ascending=False)
```

## PHASE 5 — Time & Performance Analysis

##### Monthly Revenue Trend
to show all 12 months

```python
monthly_sales = merged_df.groupby('close_month')['close_value'].sum().reset_index().sort_values(by='close_month',ascending = True)
monthly_sales = monthly_sales.reindex(range(1,13), fill_value=0)
monthly_sales
```

## PHASE 5 - KPI Engineering

#### Total Revenue KPI

```python
total_revenue = merged_df.loc[merged_df['deal_stage'] == 'Won', 'close_value'].sum()
print("Total Revenue:", total_revenue)
```

#### sales cycle length
Measures how long deals take to close

*first creating the time cycle

```python
merged_df['sales_cycle_days'] = (merged_df['close_date'] - merged_df['engage_date']).dt.days
```

```python
avg_sales_cycle = merged_df.loc[merged_df['deal_stage'] == 'Won', 'sales_cycle_days'].mean()
print("Average Sales Cycle (Days):", round(avg_sales_cycle,2))
```

#### revenue by sales agent

```python
revenue_by_agent = merged_df.loc[merged_df['deal_stage'] == 'Won'] \
                    .groupby('sales_agent')['close_value'].sum() \
                    .sort_values(ascending=False)

revenue_by_agent
```

#### which product earns most mone

```python
merged_df[merged_df['deal_stage'] == 'Won'] \
    .groupby('product')['close_value'] \
    .sum() \
    .sort_values(ascending=False)
```
